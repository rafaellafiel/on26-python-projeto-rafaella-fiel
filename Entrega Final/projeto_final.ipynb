# Considerações
Na cidade de São Paulo existe uma quantidade considerável de estudantes migrantes internacionais matriculados em distorção idade-série. Sabe-se que não é apenas a idade que deve ser considerada ao nivelar um estudante para seguir em uma série do ensino regular do Brasil. Mas, é preciso levar em conta que a forma como a nivelação tem sido feita, pode acarretar na evasão desses estudantes. Considerando que estar em um ambiente onde a faixa estária é muito diferente, causa falta de identificação em meio a turma. Além de todo o contexto de não saber a língua, estar se inserindo em uma cultura diferente e estar em situação vulnerável.
Observando os países com maiores números de estudantes em distorção, a língua pode não ser um fator tão relevante ao considerar que Angola é a terceira nacionalidade com maior índice.
