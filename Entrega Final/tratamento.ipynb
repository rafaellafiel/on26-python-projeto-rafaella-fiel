# Tratamento inicial
O presente notebook mostra como foi feito o tratamento inicial da database retirada do Portal de Imigração do INEP.
Decidi por filtrar a base de dados para o estado de SP pois a análise do Brasil inteiro, além de trabalhosa, demandaria muito da máquina que utilizo para fazer o projeto.
Posteiormente, filtrei também o ano do reporte dos dados para o mais recente, que é o ano de 2020

```python
# primeiro importando o pandas
import pandas as pd
```

```python
# lendo o csv retirado do portal de imigraçao (link no md)
df = pd.read_csv("INEP.csv", sep = ";", encoding="latin-1")
df
```

```python
# verificando informações gerais sobre a base de dados
df.info()
```

```python
# importando numpy caso necessario
import numpy as np
```

```python
# filtrando dadis aoebas do estado de SP. Segundo o IBGE, o codigo de SP é 35
sp = df.query("CO_UF in [35]")
sp
```

```python
# filtrando o que é de SP mas também apenas reportado pelo Censo 2020
sp_2020 = sp.query("NU_ANO_CENSO in [2020]")
sp_2020
```

```python
# exportando o CSV para tratamento mais aprofundado
sp_2020.to_csv("SP_INEP.csv", index=False)
```
