```python
# Nova coluna com status de idade e serie como adequada ou distorcida, comparando coluna idade da pessoa estudante e coluna de idade adequada
saopaulo.loc[saopaulo["NU_IDADE"] == saopaulo["IDADE ADEQUADA"], "STATUS"] = "ADEQUADA"
saopaulo.loc[saopaulo["NU_IDADE"] < saopaulo["IDADE ADEQUADA"], "STATUS"] = "DISTORCIDA"
saopaulo.loc[saopaulo["NU_IDADE"] > saopaulo["IDADE ADEQUADA"], "STATUS"] = "DISTORCIDA"

saopaulo
```

```python
# verificacao de dados nulos apos o tratamento e diagnosticacao do status de adequada ou distorcida
saopaulo.isnull().sum()
```

```python
# filtrando a coluna de status apenas com string adequada ou distorcida. Sao os dados relevantes para a analise
saopaulo = saopaulo[saopaulo["STATUS"].str.contains("ADEQUADA|DISTORCIDA", na=False)]
saopaulo
```
